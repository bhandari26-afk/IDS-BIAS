```python
import os
import pandas as pd
import yaml
import matplotlib.pyplot as plt
import seaborn as sns

# Load config
with open("../config.yaml", "r") as f:
    config = yaml.safe_load(f)

# Load paths from config
combined_path = "../data/combined/cicids2017_combined.csv"
preprocessed_path = "../data/processed/preprocessed.csv"


combined_path, preprocessed_path
```

```python
df_combined = pd.read_csv(combined_path)
print("Combined dataset shape:", df_combined.shape)
df_combined.head()
```

```python
import matplotlib.pyplot as plt
df_combined['Label'].value_counts().plot(kind='bar', figsize=(10,4))
plt.title("Label Distribution - Raw Combined Data")
plt.xlabel("Label")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
preprocessed_path = "../data/processed/preprocessed.csv" 
# Load preprocessed dataset
df_pre = pd.read_csv(preprocessed_path)
print("Preprocessed dataset shape:", df_pre.shape)
df_pre.head()
```

```python
print("Missing values per column:")
print(df_pre.isnull().sum().sort_values(ascending=False).head(10))

print("\nNumber of duplicate rows:", df_pre.duplicated().sum())

print("\nUnique labels:", df_pre['Label'].unique())
```

```python
df_pre['Label'].value_counts().plot(kind='bar', figsize=(10,4))
plt.title("Encoded Label Distribution - Preprocessed Data")
plt.show()
```

```python
numeric_cols = df_pre.select_dtypes('number').columns.drop('Label')
df_pre[numeric_cols].describe().T.head(10)
```

```python
numeric_df = df_pre.select_dtypes(include=['number'])

plt.figure(figsize=(12, 8))
sns.heatmap(numeric_df.corr().iloc[:20, :20], cmap="coolwarm")
plt.title("Feature Correlation Sample (first 20 numeric columns)")
plt.show()
```

```python
sample = df_pre.sample(100000, random_state=42)
sample.to_csv("../data/processed/sample_preprocessed.csv", index=False)
print("âœ… Saved sample_preprocessed.csv for quick model experiments.")
```
